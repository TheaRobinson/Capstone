```python
import pandas as pd

### Get Data
# Get Parcel ID Data
parcel_id = pd.read_csv("MKE_Parcel_ID.csv", encoding = 'latin-1', sep = ',')

# Get MPROP Data
mprop_19 = pd.read_csv("MPROP_2019.csv", encoding = 'latin-1', sep = ',')
```

```python
# Clean Data
data_19 = mprop_19[['HOUSE_NR_LO', 'SDIR', 'STREET', 'STTYPE', 'OWN_OCPD', 'P_A_LAND', 'P_A_IMPRV', 'P_A_TOTAL', 'GEO_TRACT']]
data_19['OWN_OCPD'] = data_19['OWN_OCPD'].fillna('N')
data_19['address'] = data_19['HOUSE_NR_LO'].map(str) + ' ' + data_19['SDIR'].map(str) + ' ' + \
    data_19['STREET'].map(str) + ' ' + data_19['STTYPE'].map(str)


data_19['imprv_val_share'] = data_19['P_A_IMPRV'] / data_19['P_A_TOTAL']
data_19['imprv_val_share'] = data_19['imprv_val_share'].fillna(-1)

tract_19 = data_19[['GEO_TRACT', 'imprv_val_share', 'P_A_TOTAL', 'P_A_LAND']]

tract_19 = tract_19.groupby(['GEO_TRACT']).median()

tract_19
```
